```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Liste der Symbole, die Sie abrufen möchten
# Fügen Sie hier die Tickersymbole für die zusätzlichen Tech-Aktien und Nicht-Tech-Aktien hinzu
tech_symbols = ['TSLA', 'AAPL', 'MSFT', 'GOOGL', 'AMZN']  # Tech-Aktien
non_tech_symbols = ['JPM', 'PG', 'KO', 'PFE', 'MDO']  # Nicht-Tech-Aktien
crypto_symbols = ['BTC-USD']  # Kryptowährungen

symbols = tech_symbols + non_tech_symbols + crypto_symbols

# Daten für jedes Symbol abrufen
data = yf.download(symbols, start="2020-01-01", end="2023-12-31")

# Wählen Sie nur die Schlusskurse aus
closing_prices = data['Close']

# Überprüfen Sie auf fehlende Werte
print("Missing values for each symbol:")
print(closing_prices.isnull().sum())

# Füllen Sie fehlende Werte durch Interpolation
closing_prices = closing_prices.interpolate()

# Berechnen Sie grundlegende Statistiken für jeden Datensatz
print("\nBasic statistics for each symbol:")
print(closing_prices.describe())

# Erstellen Sie Zeitreihendiagramme für jeden Datensatz
plt.figure(figsize=(14, 7))
for i in range(len(closing_prices.columns)):
    plt.plot(closing_prices.index, closing_prices.iloc[:, i], lw=2, label=closing_prices.columns[i])
plt.legend(loc='upper right', fontsize=12)
plt.ylabel('Price in $')

# Korrelationsmatrix anzeigen
correlation = closing_prices.corr()
plt.figure(figsize=(10, 10))
heatmap = sns.heatmap(correlation, annot=True, linewidths=0, vmin=-1, cmap="RdBu_r")
```

```python
import yfinance as yf
import pandas as pd
import sqlite3

# Definieren Sie die Listen der Aktiengruppen
fang_stocks = ['META', 'AMZN', 'NFLX', 'GOOGL']
bluechip_stocks = ['JNJ', 'PG', 'KO', 'MCD', 'V']
finance_stocks = ['JPM', 'GS', 'BAC']
energy_stocks = ['XOM', 'CVX', 'BP']
tech_stocks = ['TSLA', 'NVDA', 'ZM', 'SHOP']

# Alle Aktien in einer Liste zusammenführen
all_stocks = fang_stocks + bluechip_stocks + finance_stocks + energy_stocks + tech_stocks

# Daten für alle Aktien und Bitcoin abrufen
data = yf.download(all_stocks + ['BTC-USD'], start="2020-01-01", end="2023-12-31")

# Wählen Sie nur die Schlusskurse aus
closing_prices = data['Close']

# Füllen Sie fehlende Werte durch Interpolation
closing_prices = closing_prices.interpolate()

# Erstellen Sie eine SQLite-Datenbank und eine Verbindung dazu
conn = sqlite3.connect('stocks.db')

# Fügen Sie die heruntergeladenen Aktiendaten in die Datenbank ein
closing_prices.to_sql('closing_prices', conn, if_exists='replace')

def get_closing_prices(group_stocks):
    symbols = ', '.join(f'"{stock}"' for stock in group_stocks)
    query = f'SELECT Date, {symbols} FROM closing_prices'
    return pd.read_sql(query, conn, index_col='Date')

fang_prices = get_closing_prices(fang_stocks)
bluechip_prices = get_closing_prices(bluechip_stocks)
finance_prices = get_closing_prices(finance_stocks)
energy_prices = get_closing_prices(energy_stocks)
tech_prices = get_closing_prices(tech_stocks)

# Schleife über alle Aktiengruppen und berechne den Durchschnitt der Schlusskurse
for group_name, group_stocks in zip(["FANG", "Bluechip", "Finance", "Energy", "Tech"], 
                                    [fang_stocks, bluechip_stocks, finance_stocks, energy_stocks, tech_stocks]):
    query = f"SELECT Date, AVG({','.join([f'"{stock}"' for stock in group_stocks])}) as {group_name}_Avg_Close FROM closing_prices GROUP BY Date"
    avg_closing_prices = pd.read_sql_query(query, conn)
    print(f"\n{group_name} stocks average closing prices:")
    print(avg_closing_prices)
```

```python
import yfinance as yf
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt

# Define the lists of stock groups
fang_stocks = ['META', 'AMZN', 'NFLX', 'GOOGL', 'BTC-USD']
bluechip_stocks = ['JNJ', 'PG', 'KO', 'MCD', 'V', 'BTC-USD']
finance_stocks = ['JPM', 'GS', 'BAC', 'BTC-USD']
energy_stocks = ['XOM', 'CVX', 'BP', 'BTC-USD']
tech_stocks = ['TSLA', 'NVDA', 'ZM', 'SHOP', 'BTC-USD']

# Combine all stocks into a single list
all_stocks = fang_stocks + bluechip_stocks + finance_stocks + energy_stocks + tech_stocks + btc_stocks

# Retrieve data for all stocks and Bitcoin
data = yf.download(all_stocks, start="2020-01-01", end="2023-12-31")

# Select only the closing prices
closing_prices = data['Close']

# Fill missing values through interpolation
closing_prices = closing_prices.interpolate()

# Create a SQLite database and a connection to it
conn = sqlite3.connect('stocks.db')

# Insert the downloaded stock data into the database
closing_prices.to_sql('closing_prices', conn, if_exists='replace')

# Example SQL query to retrieve closing prices for a specific stock
def get_closing_prices(symbol):
    query = f'SELECT Date, "{symbol}" FROM closing_prices'
    return pd.read_sql(query, conn, index_col='Date')

# Example usage of the get_closing_prices function
fb_prices = get_closing_prices('META')

# Define a dictionary with stock groups
stock_groups = {
    'FANG': fang_stocks,
    'Blue Chip': bluechip_stocks,
    'Finance': finance_stocks,
    'Energy': energy_stocks,
    'Tech': tech_stocks,
}

# Loop over all stock groups and calculate the correlation matrix
for group_name, stocks in stock_groups.items():
    # Query the closing prices for the group of stocks
    columns = ', '.join([f'"{stock}"' for stock in stocks])
    query = f"SELECT Date, {columns} FROM closing_prices"
    group_prices = pd.read_sql_query(query, conn, index_col='Date', parse_dates=['Date'])
    # Calculate the correlation matrix for the group of stocks
    corr_matrix = group_prices.corr()
    # Plot the correlation matrix
    plt.figure(figsize=(8, 8))
    plt.imshow(corr_matrix, cmap='coolwarm', vmin=-1, vmax=1)
    plt.colorbar()
    plt.title(f'Correlation Matrix of {group_name} Stocks')
    plt.xticks(range(len(stocks)), stocks, rotation=45)
    plt.yticks(range(len(stocks)), stocks)
    
     # Add numerical values to the plot
    for i in range(len(stocks)):
        for j in range(len(stocks)):
            plt.text(i, j, round(corr_matrix.iloc[i, j], 2), ha='center', va='center', color='white')
            
            
    plt.show()
```
